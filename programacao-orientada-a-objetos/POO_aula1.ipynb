```python
class Vendedor():
    def __init__(self, nome):
        self.nome = nome
        self.vendas = 0
    
    def vendeu(self, vendas):
        self.vendas = vendas

    def bateu_meta(self, meta): #métodos 
        if self.vendas > meta:
            print(self.nome, "Bateu a meta")
        else:
            print( self.nome, "Não bateu a meta.")


Vendedor1 = Vendedor("Gabriela")
print(Vendedor1.nome)

Vendedor1.vendeu(100)
Vendedor1.bateu_meta(600)

Vendedor2 = Vendedor("João")
print(Vendedor2.nome)

```

```python
class Pessoa():
    def __init__(self, nome):
        self.nome = nome
        self.idade = 0

    def definir_idade(self, idade):
        self.idade = idade  

pessoa_1 = Pessoa("Gabriela")
print(pessoa_1.nome)

pessoa_1.definir_idade(30)
print(pessoa_1.idade)

```

```python
class Produto:
    def __init__(self, nome, preco, estoque):
        self.nome = nome
        self.preco = preco
        self.estoque = estoque 
        
    
    def vender(self, quantidade):
        if self.estoque < quantidade: 
            print("Não é pssível vender o produo")
        else:
             self.estoque = self.estoque - quantidade
            
    def repor(self, quantidade):
        if quantidade < 0:
            print("Digite um valor válido")
        else: 
            self.estoque = self.estoque + quantidade
            

acucar = Produto("açucar" ,"1.40", 12)
acucar.vender(10)

acucar.repor(-4)
acucar.estoque
acucar.repor(4)
acucar.estoque
```
